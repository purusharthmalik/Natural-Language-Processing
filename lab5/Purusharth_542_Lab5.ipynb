### 1. Write a program to get Antonyms from WordNet.

```python
from nltk.corpus import wordnet

def get_antonyms(word):
    antonyms = []
    for syn in wordnet.synsets(word):
        # Finding all the lemmas
        for lemma in syn.lemmas():
            # Checking if antonyms exist
            if lemma.antonyms():
                for ant in lemma.antonyms():
                    # Adding the antonyms to the list
                    antonyms.append(ant.name())
    return antonyms

# Printing the antonyms
for word in words:
    print(get_antonyms(word))
```

### 3. Write a program for lemmatizing words Using WordNet (Use all type of stemmers for the comparison).

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer, LancasterStemmer

def lemmatize_word(word):
    lemmatizer = WordNetLemmatizer()
    lemma = lemmatizer.lemmatize(word)
    return lemma

def stem_word(word, stemmer):
    return stemmer.stem(word)

def compare_words(words):
    stemmers = {
        "Porter": PorterStemmer(),
        "Lancaster": LancasterStemmer(),
        "Snowball": SnowballStemmer("english")
    }
    
    print("Lemmatized Words:")
    print([lemmatize_word(word) for word in words], end='\n\n')
    
    for stemmer_name, stemmer in stemmers.items():
        stemmed_words = [stem_word(word, stemmer) for word in words]
        print(f"{stemmer_name} Stemmer:")
        print(stemmed_words, end='\n\n')

compare_words(words)
```
